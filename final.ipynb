```
def get_country_code(country_name):
    for code in COUNTRIES.keys():
      if COUNTRIES[code]==country_name:
        return code
    return None
```

```

def check_login_in_year(url):
 # print(url)
  res=requests.get(url)
  soup=BeautifulSoup(res.text)
  for i in soup.select("#pageContent > div:nth-of-type(2) > div:nth-of-type(5) > div:nth-of-type(2) > ul > li:nth-of-type(4)"):
    l=(i.text).split(' ')
    if l[-3]=="year" or l[-3]=='years':
      return False
    else:
      return True
  #pageContent > div.datatable.ratingsDatatable > div:nth-child(6) > table > tbody > tr:nth-child(2) > td:nth-child(2)
```

```
def get_name_urls(soup): 
  urls =[]
  cnt=0
  b=True
  for i in soup.select("div.datatable.ratingsDatatable > div:nth-of-type(6) > table > tr"):
    if(len(i.select("img"))):
      for j in i.select("td:nth-of-type(4)"):
        url="http://codeforces.com/"+i.find('a')['href']
        if int(j.text)>= lowerbound:
          for k in i.select("img") :
            s=get_country_code(k['title'])
            if s!=None:  
              if country_to_people.get(s,False):
                country_to_people[s]+=1
              else:
                country_to_people[s]=1
          cnt+=1
        elif int(j.text) < lowerbound:
          b=False
  return cnt,b
```

```
def ith_pages_name_urls(i=1):
  url="http://codeforces.com/ratings/page/"+str(i)
  #print(url)
  cnt=0
  res=requests.get(url)
  soup=BeautifulSoup(res.text)
  temp,b=get_name_urls(soup)
  cnt+=temp
  if b:
    cnt+=ith_pages_name_urls(i+1)
  return cnt
```

```
def show(cnt):
  wm = pygal_maps_world.maps.World()
  wm.title = 'People oncodeforces whose rating over %d                total:%d people'%(lowerbound,cnt)
  wm.add('At least %d'%lowerbound, country_to_people)
  wm.render_to_file("wm.svg")
  wm.render_to_png("wm.png")
```
